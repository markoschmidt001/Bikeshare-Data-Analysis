## Investigation Overview

In my explanatory analysis i would like to focus on the trip distribution on days of the week and hour of the days. In addition I will have closer look at the ourly trip duration among user types.


## Dataset Overview

In this analysis I observed Data from the bikeshare company Ford GoBike to find patterns and answer interesting questions. For the analysis I used Januar to March 2019 from this source: ttps://s3.amazonaws.com/baywheels-data/index.html. To be able to carry out the analysis, I completed a few cleaning tasks in advance. These included creating new columns and converting data formats. I focused my analysis on the main features Duration, Starttime, Start- and Endtation and Usertype. To better understand the main features I used the following supporting features named, Day of the week, Hour of the day and Station information. I used the "Question-Visualization-Observations" framework throughout the exploration.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")

# Define a function for the repeated code to add axis label and title to all plots
def x_y_t(xL, yL, title):
    plt.title(title)
    plt.xlabel(xL)
    plt.ylabel(yL)
color = sb.color_palette()[2]
```

```python
# load in the dataset into a pandas dataframe
df_merged = pd.read_csv('ford_gobike_data_merged.csv')
```

## How are the trips distributed on the day of the weeks?

Thursday is the day of the week with the most trips taken. On weekend there are less trips taken then on weekdays.

```python
# Choose the first tuble of RGB colors
base_color = sb.color_palette()[0]

# Create a barchart
sb.countplot(data = df_merged, x='day_of_week', color=base_color, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.xticks(rotation=60)

# Use of Function to create axes and title for the plot
x_y_t('Day of Week', 'Count', 'Trip Distribution for Day of Week')
```

## How are the trips distributed on the hour of the day?

There are peek hours around 7-9 am and 16-18 pm with more trips taken. Usally trips are taken on daytime. Much less at night.

```python
# Create a barchart
sb.countplot(data=df_merged, x='our_of_day', color=base_color)

# Use of Function to create axes and title for the plot
x_y_t('Our of Day', 'Count', 'Trip Distribution for Our of Day')
```

## How is the ourly trip duration distributed among the user types?

The trip duration is in general longer at all hours of the day for customers. At night from 1-3 am it is much higher on average and from 5 to 9 am the difference between subcriber and customer is the smallest.

```python
# Create clustered bar chart
plt.figure(figsize=(10,6))
plt.title('Trip Duration to Hour of the Day', fontsize=15)
chart = sb.barplot(data=df_merged, x='our_of_day', y='duration_sec', hue='user_type', ci=None)
chart.set(xlabel='Hour of Day', ylabel='Duration (seconds)')
plt.gca().legend().set_title('');
```
